# Чтение CSV-файлов

```python
from prozhito_nlp import load_diary_from_csv

# В качестве примера будем работать с дневником Е. Шварц, который хранится в файле author_394.csv

schwartz = load_diary_from_csv("путь/к/файлу/author_394.csv") # создаем датафрейм с дневником
schwartz.head() # оцениваем отображение
```

# Очистка данных

```python
from prozhito_nlp import clean_text_column, add_year_column

# Данная функция очищает исходный текст от свойственных ему "шумов" -- нежелательных символов, следов разметки и артефактов оцифровки.

schwartz = clean_text_column(schwartz, text_column="text") # очищаем текст от шумов
schwartz = add_year_column(schwartz, date_column="date") # добавляем отдельную колонку с годом для последующих этапов анализа
schwartz.head(2) # оцениваем результаты обработки
```

# Лемматизация

```python
from prozhito_nlp import lemmatize_column

# Данная функция лемматизирует текст -- приводит все словоформы к начальной (словарной) форме.
# Это необходимый этап подготовки к количественному анализу.

schwartz = lemmatize_column(schwartz, text_column="text", new_column="tokens") # лемматизируем текст, который хранится в колонке text. создаем новую колонку tokens для лемматизированного текста.
schwartz.head(2) # оцениваем результаты обработки
```

# Базовые количественные показатели текста

```python
from prozhito_nlp import clean_punctuation, count_sentences, compute_text_statistics

# Данная функция производит расчет базовых количественных показателей текста:
# количества записей
# среднего объема записей в токенах
# общего количества токенов в дневнике
# количества уникальных токенов в дневнике
# средней длины предложений в токенах

stats = compute_text_statistics(schwartz, token_column="tokens") # подсчитываем показатели, взяв за основу колонку с лемматизированным текстом

for key, value in stats.items(): # выводим результаты
    print(f"{key}: {value}")
```

# TF-IDF

```python
from prozhito_nlp import compute_tfidf_by_year

# Данная функция подсчитывает TF-IDF по годам. Таким образом можно вывести наиболее статистически значимые слова в дневнике за определенный период (год).

schwartz_by_years = compute_tfidf_by_year( # создаем вспомогательный датафрейм, где будет храниться информация о TF-IDF дневника, и применяем функцию расчета
    df=schwartz, # указываем основной датафрейм с дневником
    text_column='tokens', # указываем колонку с лемматизированным текстом
    year_column='year', # указываем колонку с годом
    stop_words_path="путь/к/файлу/stop_words.txt", # указываем путь к файлу со стоп-словами
    top_n=20, # топ-N слов за каждый год. при значении 20 будет показывать топ-20 слов по каждому году.
    display_year=1960 # указываем год, по которому будет выведена статистика
)
```

# Визуализация TF-IDF по годам

```python
from prozhito_nlp import plot_tfidf_by_year

# Данная функция строит барчарт наиболее статистически значимых слов по годам. В выпадающем списке сверху можно выбрать интересующий год.

plot_tfidf_by_year(schwartz_by_years, top_n=20) 
```

# Поиск соответствий с тематическими словарями

```python
from prozhito_nlp import match_custom_dictionaries

# Данная функция выполняет поиск соответствий между текстом и тематическими словарями -- по темам, потенциально важным для дневниковых нарративов.
# В данной версии указаны встроенные в проект словари (в лемматизированном виде).
# Мы оставляем за пользователем право указать дополнительно собственные списки слов, но для корректной работы функции важно, чтобы они также были заранее приведены к лемматизированному виду, а их названия строились по шаблону словарь_lemm.txt

dict_path = "путь/к/папке/dictionaries/"
dict_names = [ #перечисление названий файлов, без _lemm в конце
    'phraseologisms_wiki', 'clothes', 'weather', 'health', 'items',
    'money', 'ratio', 'relatives', 'work_school', 'culture',
    'festivals', 'domestic'
]

total_matches, unique_matches = match_custom_dictionaries( # применяем функцию
    df=schwartz, # указываем основной датафрейм с дневником
    text_column='tokens_no_punkt', # указываем именно эту колонку. в ней заранее был помещен текст в лемматизированном виде, очищенный от всех знаков препинания, кроме тире
    dict_dir=dict_path,
    dict_names=dict_names,
    show_details=True
)
```

# Визуализация совпадений с тематическими словарями

```python
from prozhito_nlp import plot_total_matches, plot_matches_by_category

# Данная функция визуализирует найденные совпадения между текстом дневника и тематическими словарями.
# Предлагается два вида графиков:
# вертикальный барчарт с распределением совпадений по категориям
# и горизонтальный барчарт с перечислением наиболее частотных совпадений внутри отдельных категорий. здесь в выпадающем списке в правом верхнем углу можно выбрать интересующую категорию.

# Общее количество совпадений по категориям
plot_total_matches(total_matches)

# Наиболее частотные совпадения по категориям
plot_matches_by_category(unique_matches, schwartz, token_col='tokens_no_punkt')
```

# Лингвистические параметры

```python
from prozhito_nlp import NatashaAnalyzer, calc_percentage, analyze_verbs, analyze_pronouns, analyze_interjections, analyze_sentences, TextAnalyzer

# Данная функция производит расчет ряда лингвистических параметров текста.
# 
# Итоговый вывод содержит:
# общее количество глаголов, распределение глаголов по видам (количество + процентное соотношение), списки уникальных глаголов с разделением по виду;
# распределение глаголов по временам (общее количество + процентное соотношение), списки уникальных глаголов в разных временных формах;
# типы предложений по цели высказывания (общее количество + процентное соотношение), списки уникальных предложений по категориям (вопросительные, восклицательные, повествовательные);
# междометия (общее количество и список уникальных значений);
# общее количество местоимений, распределение местоимений по лицам (количество + процентное соотношение), общий список уникальных местоимений.


orig_text = " ".join(schwartz["text"])
lemm_text = " ".join(schwartz["tokens"])

analyzer = TextAnalyzer(orig_text, lemm_text)
analyzer.print_report()
```

# Сентимент-анализ

```python
from prozhito_nlp import load_rusentilex_dict, calculate_sentiment_score, analyze_sentiment, print_sentiment_results
from pathlib import Path

# Данная функция производит сентимент-анализ текста, основываясь на заранее очищенном словаре эмоционально-окрашенных слов RuSentiLex.
# В качестве текстового вывода получаем анализ текста по трем ключевым категориям лексики, используемой в RuSentiLex:
# оценочной лексики,
# прямым обозгачениям чувств,
# эмоционально окрашенным фактам.
# Результаты представлены по обозначенным категориям, с дополнительным делением совпадений на позитивные, нейтральные и негативные.
# Выводится общее количество совпадений с лексическими единицами категории, процент уникальных слов, совпадающих с лексиконом, а также доля совпадений относительно всей категории.
# Дополнительно выводится список всех уникальных слов, которые были обнаружены и соответствуют данной категории и сентименту.


file_path = Path("путь/к/файлу/rusentilex_clean.txt") # указываем путь к очищенному словарю RuSentiLex
lexicon = load_rusentilex_dict(file_path) # загружаем словарь RuSentiLex в виде Python-словаря

result, total_unique_words, total_category_words, schwartz = analyze_sentiment( # выполняем сентимент-анализ дневника schwartz
    schwartz, 'tokens', lexicon
)

print_sentiment_results(result, total_unique_words, total_category_words) # выводим результаты анализа
```

## График динамики сентимента во времени

```python
from prozhito_nlp import plot_sentiment_dynamics

# Данная функция визуализирует динамику сентимента во времени.
# Отдельные записи представлены на графике в виде точек, которые для наглядности соединены прямыми линиями.
# Таким образом можно не только оценить отдельные перепады настроения в тексте, но и наглядно отразить возможные разрывы в данных (иногда между записями могут быть перерывы, исчисляемые днями, месяцами и даже годами).
# По выпадающему списку в правом верхнем углу графика можно выбрать между графиком с оригинальными значениями сентимента и двумя сглаженными вариантами: фильтром Савицкого-Голея и LOWESS.
# При наведении курсора на точку, появится лейбл, отражающий выбранный временной период и числовое значение сентимента.

# Вызвать график можно несколькими способами, вот самый простой из них:
plot_sentiment_dynamics(schwartz)
```

```python
# Способ вызова графика с дополнительными настройками

fig = plot_sentiment_dynamics(
    schwartz,
    window_length=9, # длина окна
    lowess_frac=0.2, #степень сглаживания фильтра LOWESS
    # title_prefix="Сентимент в дневнике Е. Шварц" # при желании можно добавить кастомный заголовок
)
fig.show()
```
